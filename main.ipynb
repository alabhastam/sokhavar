# Analyze all the poems from three different poets
We are going to make Really regex based moshaere system. Then we are going to answer user input with a poem, which is relevant to their own choices. Then we have to know some information about the poems we are retrieving for users as outputs.

```python
from hazm import Normalizer, word_tokenize
from collections import Counter

# Initialize normalizer
normalizer = Normalizer()

# Your poets and dataset paths
files = {
    "Ø®ÛŒØ§Ù…": "/kaggle/input/khayam-poems-csv/(1).csv",
    "Ø³Ø¹Ø¯ÛŒ": "/kaggle/input/saadi-poems-csv/(1).csv",
    "Ù…ÙˆÙ„Ø§Ù†Ø§": "/kaggle/input/molana-poems-csv/.csv"
}

def top_words(path, top_n=20):
    """Return top-N most frequent words in a poet's dataset."""
    df = pd.read_csv(path)
    
    # Combine all misras into one large string
    all_text = " ".join(df["misra1"].astype(str) + " " + df["misra2"].astype(str))
    
    # Normalize Persian text (fixes 'ÙŠ'â†’'ÛŒ', 'Ùƒ'â†’'Ú©', removes diacritics, etc.)
    normalized = normalizer.normalize(all_text)
    
    # Tokenize using hazm
    tokens = word_tokenize(normalized)
    
    # Count frequencies of all tokens and return top-N
    return Counter(tokens).most_common(top_n)

# Run analysis for all poets and display results
for poet, path in files.items():
    print(f"\n Top 20 frequent words for {poet}:")
    for word, freq in top_words(path):
        print(f"{word:<10} â†’ {freq}")
```

```python
from hazm import Normalizer, word_tokenize


normalizer = Normalizer()



# === Core Function ===
def top_words_by_length(path, min_len=3, max_len=8, top_n=20):
    """Return {word_length: list of (word, freq)} for lengths in range(min_len..max_len)."""
    
    df = pd.read_csv(path)
    all_text = " ".join(df["misra1"].astype(str) + " " + df["misra2"].astype(str))
    normalized = normalizer.normalize(all_text)
    tokens = word_tokenize(normalized)


    filtered = [w for w in tokens if w.isalpha()]
    
    result = {}
    for L in range(min_len, max_len + 1):
        long_words = [w for w in filtered if len(w) == L]
        freq = Counter(long_words).most_common(top_n)
        result[L] = freq
    return result


# === Run for Each Poet ===
poet_data = {}

for poet, path in files.items():
    print(f"\nğŸ”¹ Analyzing poet: {poet}")
    length_groups = top_words_by_length(path)
    poet_data[poet] = length_groups
    
    # Display results neatly
    for L, freq_list in length_groups.items():
        print(f"\n  ğŸ“ Words with {L} letters (Top {len(freq_list)}):")
        for w, c in freq_list:
            print(f"    {w:<10} â†’ {c}")
            
```
