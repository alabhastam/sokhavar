# Load and analyze the dataset related to work 
This dataset consists of a lot of poets from Khaje-Nasir with not many poems to Rumi, who is one the most famous poets all around the world. So we can not have all of them just in one work, because we need to simplify our problem, just a little bit.


```python
import os, pandas as pd
from hazm import Normalizer

normalizer = Normalizer()

def make_csv_from_txt(root_folder, poet_name):
    beyts = []
    poet_path = os.path.join(root_folder, poet_name)
    for file in os.listdir(poet_path):
        if file.endswith(".txt"):
            text = open(os.path.join(poet_path, file), encoding="utf-8").read()
            lines = [normalizer.normalize(l.strip()) for l in text.split("\n") if l.strip()]
            for i in range(0, len(lines), 2):  # each 2 lines = one beyte
                if i+1 < len(lines):
                    beyts.append({"misra1": lines[i], "misra2": lines[i+1]})
    df = pd.DataFrame(beyts)
    df.to_csv(f"/kaggle/working/{poet_name}.csv", index=False)

# Example usage
root = "/kaggle/input/persian-poem-dataset/poems"
for poet in ["ÿÆ€åÿßŸÖ","ŸÖŸàŸÑÿßŸÜÿß","ÿ≥ÿπÿØ€å"]:
    make_csv_from_txt(root, poet)
```
